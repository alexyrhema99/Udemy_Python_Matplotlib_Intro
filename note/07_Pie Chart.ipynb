# 07_Pie Chart



```python
import matplotlib.pyplot as plt

values = [16,18,4,8]
pielabels = ['Python', 'Ruby', 'Java', 'Perl']
piecolors = ['red','orange','yellow','maroon']
pieexplode = [0,0.1,0,0]

plt.pie(values, labels=pielabels,explode=pieexplode,colors=piecolors,startangle=0,shadow=True)
plt.title('Pie Chart')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

values = [16,18,4,8]
pielabels = ['Python', 'Ruby', 'Java', 'Perl']
piecolors = ['red','orange','yellow','maroon']
pieexplode = [0,0.1,0,0]

plt.figure(figsize=(10,15)) #調整圖片大小
plt.pie(values, explode=pieexplode, colors=piecolors, startangle=0, shadow=True)
plt.title('Pie Chart')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.legend(loc='best', labels = pielabels, fontsize = 15) #用這種方式，可以讓圓餅圖widget旁邊不會有legend
plt.show()
```

## 圓餅圖的起始位置
- plt.pie(startangle)
> startangle : float, optional, default: None
>     If not *None*, rotates the start of the pie chart by *angle*
>     degrees counterclockwise from the x-axis.

```python
values = [16,18,4,8]
labels = ['Python', 'Ruby', 'Java', 'Perl']

plt.pie(values, labels=labels, startangle=0)
plt.title('Pie Chart')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

```python
values = [16,18,4,8]
labels = ['Python', 'Ruby', 'Java', 'Perl']

plt.pie(values, labels=labels, startangle=90)
plt.title('Pie Chart')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

## 調整圖片、圖例、標題大小

```python
import matplotlib.pyplot as plt

values = [16,18,4,8]
pielabels = ['Python', 'Ruby', 'Java', 'Perl']
piecolors = ['red','orange','yellow','maroon']
pieexplode = [0,0.1,0,0]

plt.figure(figsize=(10,15)) #調整圖片大小
plt.pie(values, labels=pielabels, explode=pieexplode, colors=piecolors, startangle=0, shadow=True)
plt.title('Pie Chart', fontsize = 20)
plt.xlabel('Horizontal title', fontsize = 15)
plt.ylabel('Vertical title', fontsize = 15)
plt.legend(loc='best', fontsize = 15)
plt.show()
```
