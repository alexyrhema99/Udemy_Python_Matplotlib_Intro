## 結論
這是讓圖片背景出現網格的範例程式模板

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [1,2,4,8,16]

plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(which='major',linestyle='-', color='black')
plt.minorticks_on()
plt.grid(which='minor',linestyle='--', color='gray', alpha=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [1,2,4,8,16]

plt.figure(figsize=(12,8))
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title', fontsize = 20)
plt.xlabel('Horizontal title', fontsize = 15)
plt.ylabel('Vertical title', fontsize = 15)
plt.grid(which='major',linestyle='-', color='black')
plt.minorticks_on()
plt.grid(which='minor',linestyle='--', color='gray', alpha=0.2)
plt.show()
```

## 圖片的背景增加Grid
- plt.grid(參數 line): Grid線的樣式
- plt.grid(參數 color): Grid的顏色

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

---
## 主要刻度的主線，與其之間的副線設定
- which : {'major', 'minor', 'both'}。主要刻度的主線，與其之間的副線設定。
    - The grid lines to apply the changes on.

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(which='major',linestyle='-', color='black')
plt.minorticks_on() #這一行一定要記得加
plt.grid(which='minor',linestyle='--', color='gray')
plt.show()
```

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(which='major',linestyle='-', color='blue')
plt.minorticks_on() #這一行一定要記得加
plt.grid(which='minor',linestyle='--', color='gray')
plt.show()
```

## 主線與副線的設定，是可以分開的

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(which='major',linestyle='-', color='black')
plt.minorticks_on()
plt.grid(which='minor',linestyle='--', color='gray', alpha=0.2)
plt.show()
```

## Grid只出現在x-axis或是y-axis

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(axis = 'x', which='major',linestyle='-', color='black')
plt.show()
```

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(axis = 'y', which='major',linestyle='-', color='black')
plt.show()
```

```python
plt.plot(x1, y1, 'ro--', label='students')
plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(axis = 'both', which='major',linestyle='-', color='black') #預設也是 axis = both
plt.show()
```
