## add_subplot

- add_subplot(nrows, ncols, index, **kwargs)
- add_subplot(pos, **kwargs)

*args
    Either a 3-digit integer or three separate integers describing the position of the subplot. If the three integers are *nrows*, *ncols*, and *index* in order, the subplot will take the *index* position on a grid with *nrows* rows and *ncols* columns. *index* starts at 1 in the upper left corner and increases to the right.

> * ARGS
>      一個3位整數或三個獨立的整數，用於描述子圖的位置。 如果三個整數按順序為 nrows, ncols 和 index，則子圖將採用 nrows列 和 ncols欄 的網格上的 index位置。 
> index 從左上角的1開始，向右增加。

```python
plt.rcdefaults()
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
# ax = fig.add_subplot(1,1,1, projection='3d') #意義相同
```

## 繪製多個圖

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcdefaults()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x1 = np.random.rand(100)
y1 = np.random.rand(100)
z1 = np.random.rand(100)

x2 = np.random.rand(100)
y2 = np.random.rand(100)
z2 = np.random.rand(100)

ax.scatter(x1,y1,z1,s=20, color='red', marker = "^")
ax.scatter(x2,y2,z2,s=20, color='blue', marker = "o") #s代表size

plt.show()
```

### Matplotlib對像簡介
- FigureCanvas 畫布
- Figure 圖
- Axes 坐標軸(實際畫圖的地方)

---
注意看 **13_Subplot.ipynb** 以及 **14_3d plot.ipynb** 的差異
- 主要是 使用面向對象的方式 有一點不同
- pyplot的方式中plt.subplot()參數和面向對像中的add_subplot()參數和含義都相同。

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.arange(1, 101)
print(x)

# pyplot的方式
# plt.subplot()
plt.subplot(221)
plt.plot(x, x)
 
plt.subplot(222)
plt.plot(x, -x)
 
plt.subplot(223)
plt.plot(x, x ** 2)
 
plt.subplot(224)
plt.plot(x, np.log(x))
 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.arange(1, 101)
print(x)

# 對 figure 圖
# plt.figure().add_subplot()
fig = plt.figure()
 
ax1 = fig.add_subplot(221)
ax1.plot(x, x)
 
ax2 = fig.add_subplot(222)
ax2.plot(x, -x)
 
ax3 = fig.add_subplot(223)
ax3.plot(x, x ** 2)
 
ax4 = fig.add_subplot(224)
ax4.plot(x, np.log(x))
 
plt.show()
```
