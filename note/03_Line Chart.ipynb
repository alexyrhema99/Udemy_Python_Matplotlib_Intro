## 基本操作

關於點的類型很多，沒有必要一定去記憶  
基本上就是
- 線的顏色: r = red, g = green, b = blue; 
- 線的種類: - = line, --, -., : = ...
- 點的種類: o = dot, ^ = triangle, s = square

```python
plt.plot([1,2,3,4,5], [1,2,4,6,8], 'ro') #X座標, Y座標, 顏色與點的類型。
plt.axis([0, 6, 0, 10]) #X-axis 以及 Y-axis的範圍。
plt.title("Your Title")
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,2,4,6,8], 'go') #X座標, Y座標, 顏色與點的類型。
plt.axis([0, 6, 0, 10]) #X-axis 以及 Y-axis的範圍。
plt.title("Your Title")
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,2,4,6,8], 'bo') #X座標, Y座標, 顏色與點的類型。
plt.axis([0, 6, 0, 10]) #X-axis 以及 Y-axis的範圍。
plt.title("Your Title")
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,2,4,6,8], 'b-') #X座標, Y座標, 顏色與點的類型。
plt.axis([0, 6, 0, 10]) #X-axis 以及 Y-axis的範圍。
plt.title("Your Title")
plt.show()
```

```python
plt.plot(x1, y1, 'b-') #X座標, Y座標, 顏色與點的類型。
plt.axis([0, 10, 0, 10]) #X-axis 以及 Y-axis的範圍。
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.title("Your Title")
plt.show()
```
