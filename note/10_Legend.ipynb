```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [1,2,4,8,16]
y2 = [1,1,2,3,5]

plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.subplots_adjust(left=0.3, bottom=0.3)
plt.legend(bbox_to_anchor=(-0.15, 0.17), loc='upper right')

plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

## legend 圖例的box 要對準哪個角落

在 plt.legend()函式中  
loc 在不講明的狀況下，其實是指圖例的中心要對準哪裡，預設應該是 loc = best  
因此我們可以使用 bbox_to_anchor(xposition, yposition)
> xposition, yposition 可以是 floating point

```python
x = [1,2,3]
y = [1,1,3]

px = 2
py = 2
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, '-', label='blue')
plt.plot(px, py, "o", color = "g", markersize = 10)
plt.legend(loc = 'center', fontsize = 15)

plt.title('center', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, '-', label='blue')
plt.plot(px, py, "o", color = "g", markersize = 10)
# plt.legend(loc = 'center')
plt.legend(loc = "upper left", bbox_to_anchor = (0,0), fontsize = 15)

plt.title('.loc = "upper left", bbox_to_anchor = (0,0)', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, '-', label='blue')
plt.plot(px, py, "o", color = "g", markersize = 10)
# plt.legend(loc = 'center')
plt.legend(loc = "upper right", bbox_to_anchor = (0,0), fontsize = 15)

plt.title('.loc = "upper right", bbox_to_anchor = (0,0)', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, '-', label='blue')
plt.plot(px, py, "o", color = "g", markersize = 10)
# plt.legend(loc = 'center')
plt.legend(loc = "lower left", bbox_to_anchor = (0,0), fontsize = 15)

plt.title('.loc = "lower left", bbox_to_anchor = (0,0)', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, '-', label='blue')
plt.plot(px, py, "o", color = "g", markersize = 10)
# plt.legend(loc = 'center')
plt.legend(loc = "lower right", bbox_to_anchor = (0,0), fontsize = 15)

plt.title('.loc = "lower right", bbox_to_anchor = (0,0)', fontsize = 20)
plt.show()
```

## 實際操作legend

如果把圖例放外面，則必須先擴充圖片外的留白空間

```python
x1 = [1,2,3,4,5]
y1 = [1,2,4,8,16]
y2 = [1,1,2,3,5]
```

```python
# 這是一般的圖
plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.legend(loc='best')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

```python
plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.subplots_adjust(right=0.7, bottom=0.3) 
# 擴充圖片旁邊的空白
# 下面圖片可能看不出來，但是如果用其他方法(例如視窗彈出的方式)
# 就會看出來圖片旁邊有留白了。
# 或是說整個 figsize 留白的關係，所以圖片變小了。

plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

```python
plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.subplots_adjust(right=0.7, bottom=0.3)
plt.legend(bbox_to_anchor=(1.05, 1)) #預設是用圖例的左上角對齊 bbox_to_anchor

plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

```python
plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.subplots_adjust(right=0.7, bottom=0.3)
plt.legend(bbox_to_anchor=(1.05, 0), loc = 'lower left')

plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

```python
plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.subplots_adjust(left=0.3, bottom=0.3)
plt.legend(bbox_to_anchor=(-0.15, 1.03), loc = 'upper right')

plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```

```python
plt.plot(x1, y1, 'ro-', label='students')
plt.plot(x1, y2, 'b^-', label='teachers')

plt.subplots_adjust(left=0.3, bottom=0.3)
plt.legend(bbox_to_anchor=(-0.15, 0.17), loc = 'upper right')

plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.grid(True)
plt.show()
```
