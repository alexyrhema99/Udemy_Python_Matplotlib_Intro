# 05_Fill Plot

---
## 結論
以下是將圖形填滿顏色的範例程式模板

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(0,10)
Y1 = 0.05*X*X

plt.ylim(0, 5)
plt.plot(X, Y1, color='blue')
plt.fill_between(X, Y1, color='red', alpha='0.5')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(0,10)
Y1 = 0.05*X*X
Y2 = 0.03*X*X

plt.ylim(0, 5)
plt.plot(X, Y1, color='blue')
plt.plot(X, Y2, color='red')
plt.fill_between(X, Y1, Y2, color='red', alpha='0.5')
plt.show()
```

## 填滿顏色
- plt.fill_between()
    - 參數alpha: 控制濃淡

```python
plt.ylim(0, 5)
plt.plot(X, Y1, color='blue')
plt.fill_between(X, Y1, color='blue', alpha='1') #alpha: 控制濃淡
plt.show()
```

```python
plt.ylim(0, 5)
plt.plot(X, Y1, color='blue')
plt.fill_between(X, Y1, color='blue', alpha='.5')
plt.show()
```

## 進階填滿顏色
- 將顏色填入到兩條線之間

```python
X = np.arange(0,10)
Y1 = 0.05*X*X
Y2 = 0.03*X*X
```

```python
plt.ylim(0, 5)
plt.plot(X, Y1, color='blue')
plt.plot(X, Y2, color='red')
plt.show()
```

```python
plt.ylim(0, 5)
plt.plot(X, Y1, color='blue')
plt.plot(X, Y2, color='red')
plt.fill_between(X, Y1, Y2, color='red', alpha='0.5')
plt.show()
```
