## 結論
這是 subplots 的範例程式模板

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3]
y = [1,2,3]

plt.rcdefaults()
fig = plt.figure()
fig.suptitle('Plot Title')

plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel('x1')
plt.ylabel('y1')

plt.subplot(2,2,2)
plt.plot(x,y, color='red')
plt.xlabel('x2')
plt.ylabel('y2')

plt.subplot(2,2,3)
plt.plot(x,y, color='red')
plt.xlabel('x2')
plt.ylabel('y2')

plt.subplot(2,2,4)
plt.plot(x,y, color='red')
plt.xlabel('x2')
plt.ylabel('y2')


plt.show()
```

## 開始繪製子圖

這邊雖然index變動，但是因為都只有繪製一個圖片的關係，所以電腦自動修正  
因此看不出來他實際擺的位置。
- index的順序是左上開始，由左至右，由上至下。

__subplot(nrows, ncols, index, **kwargs)__

```python
# plt.rcdefaults()
plt.subplot(2,2,1) #number of rows, number of columns, index for plot
plt.plot(x ,y)
plt.show()
```

```python
# plt.rcdefaults()
plt.subplot(2,2,2) #number of rows, number of columns, index for plot
plt.plot(x ,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3]
y = [1,2,3]

plt.rcdefaults()
fig = plt.figure()
fig.suptitle('Plot Title')

plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel('x1')
plt.ylabel('y1')

plt.subplot(2,2,2)
plt.plot(x,y, color='red')
plt.xlabel('x2')
plt.ylabel('y2')

plt.subplot(2,2,3)
plt.plot(x,y, color='red')
plt.xlabel('x2')
plt.ylabel('y2')

plt.subplot(2,2,4)
plt.plot(x,y, color='red')
plt.xlabel('x2')
plt.ylabel('y2')


plt.show()
```
