---
## 結論
這是bar chart的範例程式模板

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [1,2,4,8,16]

colors = [ 'green', 'red', 'blue', 'orange', 'lightblue']

plt.bar(x1,y1,edgecolor='black',color=colors, linewidth=2)
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

## 基本繪圖
- 水平的Bar Chart。horizental car chart: plt.barh()

```python
plt.bar(x1,y1)
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

```python
plt.barh(x1,y1)
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

---
## 柱狀圖的柱狀邊界顏色  
- plt.bar(參數 edgecolor)

```python
plt.bar(x1,y1,edgecolor='black')
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

---
## 柱狀圖的柱子顏色
- plt.bar(參數 color)
- 個別設定顏色。plt.bar(參數 colors)

```python
plt.bar(x1,y1,edgecolor='black',color="green") #green打g也可以。
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

```python
plt.bar(x1,y1,edgecolor='black',color="red") #red打r也可以。
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

```python
colors = [ 'green', 'red', 'blue', 'orange', 'white']

plt.bar(x1,y1,edgecolor='black',color=colors)
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

```python
colors = [ 'green', 'red', 'blue', 'orange', 'lightblue']

plt.bar(x1,y1,edgecolor='black',color=colors)
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```

---
## 柱狀圖的柱子邊界線
- plt.bar(參數 linewidth)

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [1,2,4,8,16]

colors = [ 'green', 'red', 'blue', 'orange', 'lightblue']

plt.bar(x1,y1,edgecolor='black',color=colors, linewidth=4.0)
plt.title('Your title')
plt.xlabel('Horizontal title')
plt.ylabel('Vertical title')
plt.show()
```
